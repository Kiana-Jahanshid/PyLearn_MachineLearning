```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
female_data = pd.read_csv("inputs/ANSUR_II_FEMALE_Public.csv" , encoding='latin1')
male_data   = pd.read_csv("inputs/ANSUR_II_MALE_Public.csv"   , encoding='latin1')

female_data.head()
```

```python
male_data.head()
# for first 5 mens , 108 numbers of information has been shown 
```

```python
# miaim 2 ta file csv ro be ham michasboonim : ----- > using ((( pd.concat )))  , it concat 2 dataframe with each other 
# in numpy we had , np.concatenate . 

concatenated_dataframes = pd.concat([female_data , male_data])
concatenated_dataframes
```

# preprocess stage : 

### 2. Preprocess dateset for converting unit of weight, unit of height and datatype of gender.

```python
# convert weight to kg 
concatenated_dataframes['weightkg'] =  concatenated_dataframes['weightkg'] / 10
concatenated_dataframes['weightkg']
```

```python
# calc heigth , convert it to cm --- > using (((  stature  )))  column 

concatenated_dataframes['stature'] = concatenated_dataframes['stature'] / 10
concatenated_dataframes['stature']
```

```python
# convert Gender column (str) ----> to number 
# female = 0 
# male   = 1
concatenated_dataframes['Gender'] = concatenated_dataframes['Gender'].replace(["Female" , "Male"] , [ 0 , 1 ])
concatenated_dataframes['Gender']
```

```python
from matplotlib.colors import ListedColormap

#plt.scatter(concatenated_dataframes['stature'] , concatenated_dataframes['weightkg'] , c=concatenated_dataframes["Gender"] ,  cmap=ListedColormap(["red" , "blue"]), marker= "o" , alpha= 0.1)
plt.scatter(concatenated_dataframes['stature'] , concatenated_dataframes['weightkg'] , c=concatenated_dataframes['Gender']==1 , marker= "o" , alpha= 0.5)
plt.scatter(concatenated_dataframes['stature'] , concatenated_dataframes['weightkg'] , c=concatenated_dataframes['Gender']==0 , marker= "o" , alpha= 0.5)

plt.legend(["men","women"])
plt.title("women & men stature and weigths ")
plt.xlabel("height(cm)")
plt.ylabel("weight(kg)")
```

```python
# test accuracy 
k = 3
test_acc_list = []
k_list = []

for i in range(15) :

    k_list.append(k)
    knn = KNN(k)
    knn.fit(X_train , Y_train)
    acc = knn.evaluate(X_test , Y_test)
    k+=2
    
    test_acc_list.append(acc)

plt.plot( k_list , test_acc_list )
plt.xlabel(" k values ")
plt.ylabel("Test Accuracy")
plt.show()
```

# 7. Calculate confusion matrix for test dataset.

```python
Y_prediction = knn.predict(X_test)
print(len(Y_prediction))
#Y_prediction
```

```python
confusion_matrix = np.zeros((2,2))

for i in range(1214) : 
    if Y_test[i] == 0 and Y_prediction[i] == 0 :   
        confusion_matrix[0][0] += 1
    elif Y_test[i] == 0 and Y_prediction[i] == 1 :    
        confusion_matrix[0][1] += 1
    elif Y_test[i] == 1 and Y_prediction[i] == 0  :    
        confusion_matrix[1][0] += 1
    elif Y_test[i] == 1 and Y_prediction[i] == 1 :     
        confusion_matrix[1][1] += 1
    
```

```python
from sklearn import metrics
confusion_matrix = metrics.confusion_matrix(Y_test, Y_prediction)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [False, True])
cm_display.plot()
plt.xlabel("predicted label")
plt.ylabel("True label")
plt.show()
```
