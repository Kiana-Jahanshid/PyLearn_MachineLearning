```python
apple_width  = 6
apple_length = 6
banana_width = 4
banana_length = 8
watermelon_width = 20
watermelon_length = 30
```

```python
n = 200
apple_width = np.random.normal(6 , 1.5  , n)
apple_length = np.random.normal(6 , 2.5  , n)

banana_width = np.random.normal(3 , 1  , n )
banana_length = np.random.normal(15 , 3.5  , n) # tedad = 200 parakendegi beine 6 , ba 3 adad var 

watermelon_width = np.random.normal(20 , 3  , n )
watermelon_length = np.random.normal(30 , 4  , n)

plt.scatter(apple_width  , apple_length , c= 'red' , marker='o' , alpha= 0.3)
plt.scatter(banana_width  , banana_length , c= 'orange' , marker='o', alpha= 0.3)
plt.scatter(watermelon_width  , watermelon_length , c= 'green' , marker='o' , alpha= 0.5)

plt.legend(["apple" , "banana" , "watermelon"])
plt.title("fruits Distribution")
plt.xlabel("width (cm)")
plt.ylabel("length (cm)")
```

```python
# labels :

# apple = 0
# banana = 1
# watermelon = 2 
```

```python
# width moz + width sib + width watermelon 
# length moz + length sib + length watermelon

width = np.concatenate((apple_width , banana_width , watermelon_width))

length = np.concatenate((apple_length , banana_length , watermelon_length))
```

```python
# be jaye inke yek array ba abade (2,600) bede , (600,2) bede

X_train = np.array([width , length]).transpose()
print(X_train.shape)
X_train


# [ fruit length , fruit width ]
#yek array 600 taeei
# har satr yek width o height ast 
# har satr yek fruit ast 
```

```python
y_apples = np.zeros(n , dtype=int)
y_apples
```

```python
y_bananas = np.ones(n , dtype=int)
y_bananas
```

```python
y_watermelon = np.ones( n , dtype=int) * 2
y_watermelon
```

```python
# LABEL HA tooye Y_train hastan -- 200 ta avali  sib  /  200 ta dovomi moz / 200 taye sevom  hendoone 
Y_train = np.concatenate((y_apples , y_bananas , y_watermelon))
Y_train
```

```python

def euclidean_distance(x1 , x2) :
    return np.sqrt(np.sum((x1 - x2)**2 ))
```

```python
s = np.array([4,5])
m = np.array([10 , 2])
euclidean_distance(s , m)
```

```python
new_fruit = np.array([7,13]) # yek mive jadid hast // test data 

# goal =  bedast avordane fasele in mive jadid az hameye mive haaye ghabli 

distances = []
for x in X_train :
    d = euclidean_distance(new_fruit , x)
    distances.append(d)

distances

# fasele mive jadidd az tamame mive ha 
```

```python
k = 5

# az beine oon 600 ta mive kodum 5 ta , be in new_fruit nazdik taran 
#pas miaim sort mikonim
```

```python
np.sort(distances)
# SATRE AVAL 5 ta nazdik tarin fasele and 
# hala oon 5 ta nemoone ei ke fasele shoon inast ro mikhaim 
# fasele ro nemikhaim , oon hamsaye ei ke fasele shon inghadre ro mikhaim 
# pas be jaye sort az , argsort estefade mikonim 
```

```python
np.argsort(distances)
# az beyne ina faghat 5 taye aval ro mikhaim 
```

```python
five_nearest_neighbours = np.argsort(distances)[0:k]
five_nearest_neighbours
# in 5 ta mive nazdik tarin mive ha be new fruit hastan 
# hala az koja tashkis bedim mive 86 om masalan sibe , ya moze , ya hendoone ?
```

```python
Y_train[five_nearest_neighbours]
```

```python
X_train[:5]
```

```python
result = np.bincount(Y_train[five_nearest_neighbours])
result
# yani 3 ta 0 , va 2 ta 1
```

```python
# hala tooye on aksariat yani oon 3 ta , miaym max ro peida mikonim 

if np.argmax(result) == 0 :
    print("apple ")
elif np.argmax(result) == 1 :
    print("banana ")
elif np.argmax(result) == 2 :
    print("watermelon ") 

# index mux number ro return mikone 
# mige oon nazdik tarin ha sib and 
```
