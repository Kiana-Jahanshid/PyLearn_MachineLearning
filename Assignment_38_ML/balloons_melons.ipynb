```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
balloons_widths  = np.random.normal( 20   , 2 , size = 600)
balloons_lengths = np.random.normal( 40   , 10 , size = 600)
balloons_weight  = abs(np.random.normal( 0.01 , 0.2 , size = 600))


melons_widths  = np.random.normal( 18 , 0.9 , size = 600)
melons_lengths = np.random.normal( 18 , 0.9 , size = 600)
melons_weight  = abs(np.random.normal( 4  , 0.8 , size = 600))

fig = plt.figure(figsize = (10,9))
ax = plt.axes(projection='3d')
ax.grid()
ax.scatter( xs = balloons_widths , ys = balloons_lengths ,  zs = balloons_weight , color= 'red'    , marker="^" , label='balloons')
ax.scatter( xs = melons_widths   , ys = melons_lengths   ,  zs = melons_weight   , color= 'green'  , marker="o" , label='melons')


plt.title('balloons and melons')
plt.xlabel("widths")
plt.ylabel("lengths")
ax.set_zlabel('weights(kg)',  labelpad=-5)
plt.legend()
plt.show()
```
